# IX) Gestion des valeurs aberrantes  

Nous créons deux nouvelles variables :  
L'une regroupe les colonnes ayant le tag '_100g'_  
L'autre regroupe les deux exeptions à la première : satured_fat est inclus dans fat, idem pour sugar dans carbohydrates.

Remplacer les valeurs aberrantes par NaN:  
1) Les valeurs au-dessus de 100 et en dessous de 0 ne sont pas admises par produit.  
2) La valeur de graisses saturés / aux graisses, et celles de sodium / sel ne doivent pas dépasser. 


```python
Colonnes_100g = Data_Open_Food_4[["fat_100g","sugars_100g","proteins_100g","salt_100g","sodium_100g","fruits-vegetables-nuts-estimate-from-ingredients_100g","saturated-fat_100g","carbohydrates_100g"]]
Colonnes_100g.describe()
```

```python
for i in Colonnes_100g.columns :
        Colonnes_100g[i] = Colonnes_100g[i].apply(lambda x: np.nan if x > 100 else x)
        Colonnes_100g[i] = Colonnes_100g[i].apply(lambda x: np.nan if x < 0 else x)
      
        
# https://stackoverflow.com/questions/38579532/pandas-equivalent-of-np-where
```
