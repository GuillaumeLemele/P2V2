### Fonctions utilitaires : représentation des données manquantes

```python
# Fonction utilitaire pour Missingno.
# Source : https://github.com/ResidentMario/missingno/blob/master/missingno/utils.py

def nullity_filter(df, filter=None, p=0, n=0):
 if filter == 'top':
     if p:
            df = df.iloc[:, [c >= p for c in df.count(axis='rows').values / len(df)]]
     if n:
            df = df.iloc[:, np.sort(np.argsort(df.count(axis='rows').values)[-n:])]
 elif filter == 'bottom': 
     if p:
            df = df.iloc[:, [c <= p for c in df.count(axis='rows').values / len(df)]]
     if n:
            df = df.iloc[:, np.sort(np.argsort(df.count(axis='rows').values)[:n])]
 return df
```

### Fonctions utilitaires : imputation des valeurs manquantes

```python
# Imputation itérative sur le df avec enable_iterative_imputer et IterativeImputer de scikit-learn.
# Crée une copie du DataFrame d'origine en utilisant la méthode copy() pour protéger le DataFrame d'origine.
# Enfin, la fonction renvoie le DataFrame modifié avec les colonnes imputées.


def impute_columns(df, cols):
    from sklearn.experimental import enable_iterative_imputer
    from sklearn.impute import IterativeImputer

    df_imputed = df.copy()
    iterimp = IterativeImputer()
    # Transformer that imputes missing values by modeling each feature with missing values as a function of other features in a round-robin fashion.

    for col in cols:
        df_imputed[col] = iterimp.fit_transform(df_imputed[[col]])
    # The function iterates over the list of column names ; fit the imputer to the specified column &; transform the column by replacing missing values with imputed values.

    return df_imputed
```
